```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt

#Função para calcular cosseno:
def func_cos(x, n):
#Usa a fórmula de Taylor calculada para cosseno:
  cos_app = 0
  for i in range(n):
    coeficiente = (-1)**i
    denominador = math.factorial(2*i)
    numerador = x**(2*i)
    cos_app += (coeficiente)*((numerador)/(denominador))
    
  return cos_app

#Calculo dos items para formar o grafico
angulos = np.arange(-2*np.pi,2*np.pi,0.1)
chama_cos = np.cos(angulos)
fig, ax = plt.subplots()
ax.plot(angulos,chama_cos)

for i in range(1,5):
    cos = [func_cos(j,i) for j in angulos]
    ax.plot(angulos,cos)

ax.set_ylim([-7,4])

#Criação de legenda para o gráfico
legenda = ['Função cosseno']
for i in range(1,5):
    #Legenda da série de n=0 a n=3
    legenda.append(f'Série de Taylor - n = {i - 1}')
ax.legend(legenda, loc=3)

plt.show()
```

```python
#b) Na questão b ele pede para calcula f ate o grau 6 ou seja até T3
x = [np.pi/4, np.pi/2, np.pi]

for n in arange(1, 5):
  for i in arange(3):
    print(f"T{n-1}({x[i]}) = {func_cos(x[i], n)}")
  print("\n")

for i in arange(3):
    print(f"f({x[i]}) = {np.cos(x[i])}")
```
